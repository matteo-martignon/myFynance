```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from datetime import datetime

now = datetime.utcnow().strftime("%Y%m%d%H%M%S")
```

```python
def clean_float(x):
    l = str(x).replace(" ", "").replace(",", ".").split(".")
    if len(l) == 1:
        return l[0]
    else:
        return "".join(l[:-1]) + "." + l[-1]
```

```python
pd.to_datetime(now)
```

```python
path = "data/data.csv"
```

```python
df = pd.read_csv(path, sep=";", parse_dates=["Data"], index_col="Data")
```

```python
df["Importo"] = df.Importo.apply(clean_float).astype(np.float64)
```
